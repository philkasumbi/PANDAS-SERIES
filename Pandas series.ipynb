# GROUP BY AND AGGREGATING

```python
df5 = pd.read_csv("/home/philip-kasumbi/Desktop/Pandas_test/Flavors.csv")
```

```python
df5
```

```python
group_by_frame = df5.groupby('Base Flavor')
```

```python
group_by_frame.mean('Flavor Rating','Texture Rating','Total Rating')
```

```python
df5.groupby('Base Flavor').mean('Flavor Rating','Texture Rating','Total Rating')
```

```python
df5.groupby('Base Flavor').count()
```

```python
df5.groupby('Base Flavor').min()
```

```python
df5.groupby('Base Flavor').max()
```

```python
df5.groupby('Base Flavor').sum()
```

```python
df5.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','min','sum','count'],'Texture Rating':['mean','max','min','sum','count']})
```

```python
df5.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','min','sum','count']})
```

```python
df5.groupby('Base Flavor').describe()
```

# MERGING,JOINING AND CONCATENATING DATAFRAMES

```python
df1 = pd.read_csv('/home/philip-kasumbi/Desktop/Pandas_test/LOTR.csv')
df1
```

```python
df2 = pd.read_csv('/home/philip-kasumbi/Desktop/Pandas_test/LOTR 2.csv')
```

```python
df2
```

```python
df1.merge(df2,how = 'inner', on = ['FellowshipID','FirstName'])
```

```python
df1.merge(df2,how = 'outer')
```

```python
df1.merge(df2,how = 'left')
```

```python
df1.merge(df2,how = 'right')
```

```python
df1.merge(df2,how = 'cross')
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```
